```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_table('code/01-university-eligibility/maturitaetsquoten.tsv')
```

```python
df.set_index(['Kanton', 'Geschlecht'], inplace=True)
```

```python
df.head()
```

```python
ten_years = df.groupby('Kanton').mean().sort_values(
  '10_years_difference').transpose().loc[['2006','2016', 
                                                         '10_years_difference']]
```

```python
df['10_years_difference'] = df['2016'] - df['2006']
```

## Comparison of the biggest loosers and winners

```python

ten_years.loc[['2006', '2016'],['OW', 'SH', 'UR', 'AG', 'BL']].plot()
```

## Men vs Women today

```python
df.loc[:,['2016']].plot(kind='barh', figsize=(4,10))
# would be nice to have this sorted, but ... I guess this is not working with
#  MultiIndices? Must do more research
```

```python
# more noodling around with the code
df.stack().reset_index()
```

```python
# Per suggestion of Patrice: calculate difference between men and women
# Slicing as per http://stackoverflow.com/questions/14964493/multiindex-based-indexing-in-pandas
All = slice(None)
df.loc[(All,'Männer'),['2016']] - df.loc[(All, 'Frauen'),['2016']]
```

```python
# alternative version using cross section
df.xs('Männer', level=1)['2016'] - df.xs('Frauen', level=1)['2016']

# this gives me actual values, which is nice
```

```python
delta_df = pd.DataFrame(
  df.xs('Männer', level=1)['2016'] - df.xs('Frauen', level=1)['2016'], 
  columns=['2016'])
delta_df
```

```python
delta_df['1980'] = df.xs('Männer', level=1)['1980'] - df.xs('Frauen', level=1)[
  '1980']
```

```python
delta_df
```

```python
plt.style.use('stylelib/matplotlibrc')

delta_df.sort_values('2016').plot(kind='barh', figsize=(4,10))
```

```python
delta_df.describe()
```

So, what does this show us? While in 1980, the difference between genders was lower, but tended towards more men getting the qualification.

Today, it is way more women that are getting qualified for higher education, with a maximum difference of almost 22 percent points in one case.
